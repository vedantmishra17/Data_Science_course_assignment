```python
# taking input from the user
n1=int(input('enter the first: '))
n2=int(input('enter the second: '))
l=[]
for gcd in range(1,n2+1):
  if n1 % gcd==0 and n2 % gcd==0:
    l.append(gcd)
l.sort()
print(f'gcd of {n1} and {n2} is {l[-1]}')
```

31. Write a program to generate a list of prime numbers using the Sieve of Eratosthenes algorithm and list
comprehension.


```python
n = int(input("Enter an upper limit for prime numbers: "))

# Create a boolean list "is_prime" and initialize all entries as True
is_prime = [True] * (n+1)
is_prime[0] = is_prime[1] = False  # 0 and 1 are not prime

for p in range(2, int(n**0.5) + 1):
    if is_prime[p]:
        # Mark all multiples of p as non-prime
        for i in range(p*p, n+1, p):
            is_prime[i] = False

# Use list comprehension to generate the list of prime numbers
prime_list = [p for p in range(2, n+1) if is_prime[p]]

print("Prime numbers up to", n, "are:")
print(prime_list)
```

35. Create a program that generates Pascal's triangle up to a specified number of rows using list
comprehension.


```python
rows = int(input("Enter the number of rows for Pascal's triangle: "))

if rows <= 0:
    print("Invalid input. Please enter a positive integer.")
else:
    # Initialize Pascal's triangle with the first row
    triangle = [[1]]

    # Generate and print Pascal's triangle
    for i in range(1, rows):
        # Generate the next row using list comprehension
        next_row = [1] + [triangle[i-1][j-1] + triangle[i-1][j] for j in range(1, i)] + [1]
        triangle.append(next_row)

    # Print Pascal's triangle
    for row in triangle:
        print(" ".join(map(str, row)))
```

37. Write a program that finds the longest word in a sentence using list comprehension.


```python
# Input: Get a sentence from the user
sentence = input("Enter a sentence: ")

# Use list comprehension to split the sentence into words
words = sentence.split()

# Find the longest word using list comprehension
longest_word = max(words, key=lambda word: len(word))

# Print the longest word
print(f"The longest word in the sentence is: {longest_word}")

```

38. Create a program that filters a list of strings to include only those with more than three vowels using list
comprehension.


```python
# Function to calculate the factorial of a number
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

# List comprehension to calculate the factorial of numbers from 1 to 5
factorials = [factorial(n) for n in range(1, 6)]

# List comprehension to calculate the sum of digits for each factorial
sum_of_digits = [sum(int(digit) for digit in str(factorial)) for factorial in factorials]

# Print the results
for n, factorial, digit_sum in zip(range(1, 6), factorials, sum_of_digits):
    print(f"The factorial of {n} is {factorial} and the sum of its digits is {digit_sum}")
```

40. Write a program that generates a list of prime palindromic numbers using list comprehension.

```python
# Function to check if a number is prime
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True

# List comprehension to generate prime palindromic numbers
prime_palindromic_numbers = [num for num in range(1, 1000) if is_prime(num) and str(num) == str(num)[::-1]]

# Print the list of prime palindromic numbers
print("Prime Palindromic Numbers:")
print(prime_palindromic_numbers)
```
