Problem 1: Bank Account Create a class representing a bank account with attributes like account number, account holder name, and balance. Implement methods to deposit and withdraw money from the account.

```python
class BankAccount:
    def set_account_details(self, account_number, account_holder_name, balance=0.0):
        self.account_number = account_number
        self.account_holder_name = account_holder_name
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            return f"Deposited Rs{amount:.2f}. New balance: Rs{self.balance:.2f}"
        else:
            return "Invalid deposit amount. Amount must be greater than 0."

    def withdraw(self, amount):
        if amount > 0:
            if amount <= self.balance:
                self.balance -= amount
                return f"Withdrew RS{amount:.2f}. New balance: Rs{self.balance:.2f}"
            else:
                return "Insufficient balance."
        else:
            return "Invalid withdrawal amount. Amount must be greater than 0."

    def get_balance(self):
        return f"Account balance for {self.account_holder_name}: Rs{self.balance:.2f}"

    def __str__(self):
        return f"Account Number: {self.account_number}\nAccount Holder: {self.account_holder_name}\nBalance: Rs{self.balance:.2f}"


account = BankAccount()
account.set_account_details("12345", "John Doe")
print(account)
print(account.deposit(1000))
print(account.withdraw(500))
print(account.get_balance())
```

Problem 4: Library Catalog Create classes representing a library and a book. Implement methods to add books to the library, borrow books, and display available books.



```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.is_available = True

    def __str__(self):
        return f"Title: {self.title}\nAuthor: {self.author}\nStatus: {'Available' if self.is_available else 'Borrowed'}"


class Library:
    def __init__(self):
        self.catalog = {}

    def add_book(self, book):
        if book.title not in self.catalog:
            self.catalog[book.title] = book
        else:
            print("Book with the same name already exists in the catalog.")

    def borrow_book(self, title,hello):
        if title in self.catalog:
            book = self.catalog[title]
            if book.is_available:
                book.is_available = False
                return f"You have borrowed the book: {book.title}"
            else:
                return "Sorry, the book is already borrowed."
        else:
            return "Book with the provided name is not in the catalog."

    def display_available_books(self):
        available_books = [book for book in self.catalog.values() if book.is_available]
        if available_books:
            for book in available_books:
                print(book)
        else:
            print("No books are available at the moment.")
book = Book('Rashmirathi', 'Ramdhari Singh Dinkar')
library = Library()
library.add_book(book)  # Pass the Book instance to the add_book method
library.display_available_books()
```

Problem 6: Shape Calculation Create a class representing a shape with attributes like length, width, and height. Implement methods to calculate the area and perimeter of the shape.


```python
class Shape:
    def __init__(self, length, width, height=0, base=0, sides=0):
        self.length = length
        self.width = width
        self.height = height
        self.base = base
        self.sides = sides

    def area_calculate(self):
        if self.sides == 3:
            return 0.5 * self.base * self.height
        elif self.sides == 4:
            if self.length == self.width:
                return self.length * self.width
            else:
                return self.length + self.width
        else:
            return None

    def calculate_perimeter(self):
        if self.sides == 3:
            return self.length + self.width + self.base
        elif self.sides == 4:
            if self.length == self.width:
                return 4 * self.length
            else:
                return 2 * (self.length + self.width)
        else:
            return None

shapes = Shape(12, 13, sides=4)
print(shapes.area_calculate())
print(shapes.calculate_perimeter())
```

Problem 10: ToDo List Create a class representing a ToDo list with attributes like tasks and due dates. Implement methods to add tasks, mark tasks as completed, and display pending tasks.

```python
class ToDoList:
    tasks=[]
    due_dates=[]
    def add_task(self, task, due_date):
        self.tasks.append(task)
        self.due_dates.append(due_date)
        return 'task added to todo list'
    def mark_task_as_completed(self, task):
        self.tasks.remove(task)
        return f'removing the {self.task} task as it is completed '
    def display_pending_tasks(self):
        for task, due_date in zip(self.tasks, self.due_dates):
            print(f"Task: {task}")
            print(f"Due Date: {due_date}")

todo=ToDoList()
todo.add_task('complete_assignment','12/9/2023')
todo.display_pending_tasks()
```
