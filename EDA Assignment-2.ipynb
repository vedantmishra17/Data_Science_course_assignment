7. What are the most popular product reviews?

So we have different review for each product and we cannot do anything with that so we have to modifiy it so i have written a code to extract postive words and and flag it a good review other wise it is a neutral or bad review this way it will be easy for us to see what are the popular reviews

```python
positive_words = ['good', 'great', 'excellent', 'awesome']
negative_words = ['bad', 'poor', 'terrible', 'awful']
neutral_words = ['okay', 'fine', 'satisfactory', 'average']

# Function to label review titles
def label_review(title):
    for word in positive_words:
        if word in title.lower():
            return 'Good Review'
    for word in negative_words:
        if word in title.lower():
            return 'Bad Review'
    for word in neutral_words:
        if word in title.lower():
            return 'Neutral Review'
    return 'Unclassified Review'

# Apply the function to create a new column
df['Review_Label'] = df['review_title'].apply(label_review)

```

```python
df['Review_Label'].value_counts().plot(kind='bar')
```

As per our code the most popular reviews are good reviews

The dataset titled "Spotify Data: Popular Hip-hop Artists and Tracks" provides a curated collection of
approximately 500 entries showcasing the vibrant realm of hip-hop music. These entries meticulously compile
the most celebrated hip-hop tracks and artists, reflecting their significant influence on the genre's landscape.
Each entry not only highlights the popularity and musical composition of the tracks but also underscores the
creative prowess of the artists and their profound impact on global listeners.

**Column Descriptors:**


**Artist:** The name of the artist, providing direct attribution to the creative mind behind the track.

**Track Name:** The title of the track, encapsulating its identity and essence.

**Popularity:** A numeric score reflecting the track's reception and appeal among Spotify listeners.

**Duration (ms):** The track's length in milliseconds, detailing the temporal extent of the musical experience.

**Track ID:** A unique identifier within Spotify's ecosystem, enabling direct access to the track for further
exploration.

```python
# converting the duration columns in minutes
from datetime import timedelta
def milliseconds_to_minute(milliseconds):
  minutes = round(timedelta(milliseconds=milliseconds).total_seconds() / 60,2)
  return minutes
```

```python
df2['Duration']=df2['Duration (ms)'].apply(milliseconds_to_minute)
```

```python
# dropping the duration in ms column
df2.drop(columns=['Duration (ms)'], inplace=True)
```

4. Identify the top 5 longest songs among the tracks of the top 5 popular artists.Represent the duration of the top 5 longest songs among the tracks of the top 5 popular artists using BarPlot.




```python
df2.head()
```

```python
artist_popularity = df2.groupby('Artist')['Popularity'].mean()
top_5_artists = artist_popularity.nlargest(5)
top_tracks = df2[df2['Artist'].isin(top_5_artists.index)]
top_5_longest_songs = top_tracks.nlargest(5, 'Duration')

# Plotting the duration of the top 5 longest songs
plt.figure(figsize=(10, 6))
plt.bar(top_5_longest_songs['Track Name'], top_5_longest_songs['Duration'], color='skyblue')
plt.title('Duration of Top 5 Longest Songs Among Tracks of Top 5 Popular Artists')
plt.xlabel('Track Name')
plt.ylabel('Duration (in seconds)')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```
